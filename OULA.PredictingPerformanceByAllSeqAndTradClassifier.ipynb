```python
import os
def readDir(dirPath):
    if dirPath[-1] == '/':
        print(u'errpath')
        return
    allFiles = []
    if os.path.isdir(dirPath):
        fileList = os.listdir(dirPath)
        for f in fileList:
            f = dirPath+'/'+f
            if os.path.isdir(f):
                subFiles = readDir(f)
                allFiles = subFiles + allFiles 
            else:
                allFiles.append(f)
        return allFiles
    else:
        return 'Error,not a dir'
    
def props_to_onehot(props):
    props=props.detach().cpu().numpy()
    a=np.argmax(props,axis=1)
    b=np.zeros((len(a),props.shape[1]))
    b[np.arange(len(a)),a]=1
    return b
```

```python
#一些训练的参数
test_percent=0.25

allfiles=readDir('Enc\\Seq.ALL')
allfiles.sort()
res_Prediction=[]
res_Prediction.append(['Course','Classifier','Acc','Pre','Rec','F1'])
for file in allfiles:
    course_name=file[12:21]
    res_thisCourse=[]
    res_thisCourse.append(course_name)
    
    df=pd.read_csv(file)
    x=df.iloc[:,0:-1]
    x=x.values
    targets=df.iloc[:,-1]
    ls_x,ls_test_x,ls_y,ls_testy = train_test_split(x,targets,test_size=test_percent)
    
    #NB
    res_thisCourse.append('NaiveBayes')
    model=sk.naive_bayes.MultinomialNB()
    model.fit(ls_x,ls_y)
    predRes=model.predict(ls_test_x)
    res_thisCourse.append(sk.metrics.accuracy_score(ls_testy,predRes))
    res_thisCourse.append(sk.metrics.precision_score(ls_testy,predRes,average='macro'))
    res_thisCourse.append(sk.metrics.recall_score(ls_testy,predRes,average='macro'))
    res_thisCourse.append(sk.metrics.f1_score(ls_testy,predRes,average='macro'))
    res_Prediction.append([
        course_name,
        'NaiveBayes',
        sk.metrics.accuracy_score(ls_testy,predRes),
        sk.metrics.precision_score(ls_testy,predRes,average='macro'),
        sk.metrics.recall_score(ls_testy,predRes,average='macro'),
        sk.metrics.f1_score(ls_testy,predRes,average='macro')
    ])

    #KNN
    model=sk.neighbors.KNeighborsClassifier()
    model.fit(ls_x,ls_y)
    predRes=model.predict(ls_test_x)
    res_Prediction.append([
        course_name,
        'KNN',
        sk.metrics.accuracy_score(ls_testy,predRes),
        sk.metrics.precision_score(ls_testy,predRes,average='macro'),
        sk.metrics.recall_score(ls_testy,predRes,average='macro'),
        sk.metrics.f1_score(ls_testy,predRes,average='macro')
    ])
    
    #SVM
    model=sk.svm.LinearSVC()
    model.fit(ls_x,ls_y)
    predRes=model.predict(ls_test_x)
    res_Prediction.append([
        course_name,
        'SVM',
        sk.metrics.accuracy_score(ls_testy,predRes),
        sk.metrics.precision_score(ls_testy,predRes,average='macro'),
        sk.metrics.recall_score(ls_testy,predRes,average='macro'),
        sk.metrics.f1_score(ls_testy,predRes,average='macro')
    ])
    
    #RF
    model=sk.ensemble.RandomForestClassifier()
    model.fit(ls_x,ls_y)
    predRes=model.predict(ls_test_x)
    res_Prediction.append([
        course_name,
        'RandomForest',
        sk.metrics.accuracy_score(ls_testy,predRes),
        sk.metrics.precision_score(ls_testy,predRes,average='macro'),
        sk.metrics.recall_score(ls_testy,predRes,average='macro'),
        sk.metrics.f1_score(ls_testy,predRes,average='macro')
    ])
    
    #LogR
    model=sk.linear_model.LogisticRegression()
    model.fit(ls_x,ls_y)
    predRes=model.predict(ls_test_x)
    res_Prediction.append([
        course_name,
        'LogR',
        sk.metrics.accuracy_score(ls_testy,predRes),
        sk.metrics.precision_score(ls_testy,predRes,average='macro'),
        sk.metrics.recall_score(ls_testy,predRes,average='macro'),
        sk.metrics.f1_score(ls_testy,predRes,average='macro')
    ])

dfRes=pd.DataFrame(res_Prediction)
dfRes.to_csv("OULA.Exp.Res\\ALL.Seq.Traditional.Classifier.csv",index=False)
print('ALL Completed!')
```
